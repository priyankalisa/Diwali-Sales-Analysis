```python
import pandas as pd

file_path = "/content/drive/MyDrive/Diwali Sales Data.csv"
df = pd.read_csv(file_path,encoding="latin1")
print(df.shape)
df.head()
```

```python
print(df.info())
```

```python
# drop unrelated/blank columns
df.drop(["Status","unnamed1"],axis=1,inplace=True)
print(df.isnull().sum())
```

```python
print(df.info())
```

```python
print(df.isnull())
```

```python
print(df.isnull().sum())
```

```python
df.dropna(axis=0,inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df["Amount"]=df["Amount"].astype(int)
print(df.info())
```

```python
df.rename(columns={"Marital_Status": "shaadi"})
```

```python
print(df[["Age","Orders","Amount"]].describe())
```

```python
# Which gender spent more on average during Diwali?

# Bar chart comparing Male vs Female average purchase amount.
import matplotlib.pyplot as plt
import pandas as pd
avg_amount=df.groupby("Gender")["Amount"].mean()
print(avg_amount)
plt.bar(avg_amount.index,avg_amount.values,color=["red","blue"])
plt.xlabel("Gender")
plt.ylabel("Average Amount")
plt.title("Average Spending by Gender during Diwali")
plt.show()
```

```python
# Which age group contributes the most to sales?

# Bar chart with Age group vs Total purchase amount.
age_group=df.groupby("Age")["Amount"].sum()
print(age_group)
plt.bar(age_group.index,age_group.values,edgecolor="black")
plt.xlabel("Age Group")
plt.ylabel("Total Purchase Amount")
plt.title("Sales Contribution by Age Group during Diwali")
plt.show()
```

```python
# Which marital status (Married / Unmarried) spent more overall?
# → Group by Marital_Status and plot bar chart.
married_unmarried=df.groupby("Marital_Status")["Amount"].sum()
print(married_unmarried)
plt.bar(married_unmarried.index,married_unmarried.values,color=["red","blue"])
plt.xlabel("Marital Status")
plt.ylabel("Amount")
plt.title("Which marital status (Married / Unmarried) spent more overall?")
plt.xticks([0,1],["Unmarried","Married"])
plt.show()
```

```python
# What’s the spending pattern based on occupation?
# → Bar chart / pie chart showing Occupation vs Total Purchase.
occupation_amount=df.groupby("Occupation")["Amount"].sum()
print(occupation_amount)
fig,ax=plt.subplots(1,2,figsize=(10,5))
plt.suptitle("Spending Pattern Based on Occupation")
ax[0].barh(occupation_amount.index,occupation_amount.values)
ax[0].set_ylabel("Occupation")
ax[0].set_xlabel("Total Purchase Amount")
ax[0].set_title("Bar Chart")
ax[1].pie(occupation_amount.values,labels=occupation_amount.index,autopct='%1.1f%%')
ax[1].set_title("Pie Chart")
fig.tight_layout()
plt.show()
```

```python
# Which state has the highest sales and number of orders?
# → Horizontal bar chart for top 10 states.
result=df.groupby("State")[["Amount","Orders"]].sum().sort_values(by="Amount",ascending=False).head(10)
print(result)
plt.barh(result.index,result["Amount"])
plt.ylabel("State")
plt.xlabel("Amount")
plt.title("Top 10 States by Sales")
plt.show()
```

```python
# Which product category generated the highest revenue?
# → Bar chart for Product_Category vs Purchase.
df["revenue"]=df["Orders"]*df["Amount"]
result=df.groupby("Product_Category")["revenue"].sum()
plt.barh(result.index,result.values)
plt.ylabel("Product Category")
plt.xlabel("Revenue")
plt.title("Revenue by Product Category")
print(result)
```

```python
# Top 10 most sold products during Diwali?
# → Bar chart sorted by product counts.
result=df.groupby("Product_Category")["Orders"].sum().sort_values(ascending=False).head(10)
print(result)
plt.barh(result.index,result.values)
plt.ylabel("Product Category")
plt.xlabel("Product Count")
plt.title("Top 10 most sold products during Diwali")
plt.show()
```

```python
# Which product category is more popular among men vs women?
# → Stacked Bar chart (Gender vs Product_Category).
result = df.groupby(["Product_Category","Gender"])["Orders"].sum().reset_index()
pivot_df = result.pivot(index="Product_Category", columns="Gender", values="Orders")
print(pivot_df)
pivot_df.plot(kind="bar", stacked=True, figsize=(10,6))
plt.title("Product Category Popularity by Gender (Stacked)")
plt.ylabel("Number of Orders")
plt.show()
```

```python
# Average purchase amount by age group for each product category
# → Grouped bar chart with Age_Group on X-axis, Avg Purchase on Y-axis.
result = df.groupby(["Age Group","Product_Category"])["Amount"].mean().reset_index()
print(result.head())
plt.bar(result["Age Group"],result["Amount"])
plt.xlabel("Age Group")
plt.ylabel("Average Purchase Amount")
plt.title("Average purchase amount by age group")
plt.show()
```
